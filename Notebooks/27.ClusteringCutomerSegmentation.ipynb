# Clustering

## Library Imports

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Load Data

```python
data = pd.read_csv('Dataset/\Mall_customers.csv', index_col='CustomerID')
data.columns
```

```python
data.columns = ['Gender', 'Age', 'AanualIncome', 'SpendingScore']
data.head()
```

## EDA

```python
df = data.copy()
df.info()
```

```python
gender_dist = df.Gender.value_counts()
gender_dist.index, gender_dist.values
```

```python
sns.set_style('darkgrid')
sns.barplot(x=gender_dist.index, y=gender_dist.values)
```

## Pre-processing

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
encoder = LabelEncoder()
df.Gender = encoder.fit_transform(df.Gender)
df.head()
```

```python
scaler = StandardScaler()
df[['Age', 'AanualIncome', 'SpendingScore']] = scaler.fit_transform(df[['Age', 'AanualIncome', 'SpendingScore']])
df.head()
```

## KMeans

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=33)
df['cluster'] = kmeans.fit_predict(df[['AanualIncome', 'SpendingScore']])
df.head()
```

```python
sns.scatterplot(data=df, x='AanualIncome', y='SpendingScore', hue='cluster')
```

```python
kmeans.inertia_
```

```python
# Elbow Method
inertia = []
n_clusters = []
for k in range(2, 10):
    model = KMeans(n_clusters=k)
    model.fit(df[['AanualIncome', 'SpendingScore']])
    inertia.append(model.inertia_)
    n_clusters.append(k)  
```

```python
sns.lineplot(x=n_clusters, y=inertia)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.n_clusters
```

```python
set(kmeans.labels_)
```

```python
# validate
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
sscore = silhouette_score(df[['AanualIncome', 'SpendingScore']], df.cluster)
sscore
```

```python
ssamples = silhouette_samples(df[['AanualIncome', 'SpendingScore']], df.cluster)
ssamples
```

```python
# Visualization
sns.set_style('whitegrid')
ylow = 10
for c in set(kmeans.labels_):
    csample = ssamples[df.cluster==c]
    csample.sort()
    yup = ylow+csample.shape[0]
    plt.fill_betweenx(np.arange(ylow, yup), 0, csample)
    ylow = yup + 10
plt.show() 
```

## Agglomerative hierarchical clustering

```python
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=5)
```

```python
df['cluster'] = agg.fit_predict(df[['AanualIncome', 'SpendingScore']])
```

```python
silhouette_score(df[['AanualIncome', 'SpendingScore']], df.cluster)
```

```python
sns.set_style('whitegrid')
ssamples = silhouette_samples(df[['AanualIncome', 'SpendingScore']], df.cluster)
ylow = 10
for c in set(agg.labels_):
    csample = ssamples[df.cluster==c]
    csample.sort()
    yup = ylow+csample.shape[0]
    plt.fill_betweenx(np.arange(ylow, yup), 0, csample)
    ylow = yup + 10
plt.show() 
```

```python
for k in range(2,7):
    agg = AgglomerativeClustering(n_clusters=k, linkage='ward')
    df['cluster'] = agg.fit_predict(df[['AanualIncome', 'SpendingScore']])
    sscore = silhouette_score(df[['AanualIncome', 'SpendingScore']], df.cluster)
    ssamples = silhouette_samples(df[['AanualIncome', 'SpendingScore']], df.cluster)
    ylow = 10
    for c in set(agg.labels_):
        csample = ssamples[df.cluster==c]
        csample.sort()
        yup = ylow+csample.shape[0]
        plt.fill_betweenx(np.arange(ylow, yup), 0, csample)
        ylow = yup + 10
    plt.title(f'Score: {sscore: .2f} | n_clusters: {k}')
    plt.show() 
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
for eps in np.linspace(0.4, 0.6, 10):
    dbscan = DBSCAN(eps=eps, min_samples=8)
    df['cluster'] = dbscan.fit_predict(df[['AanualIncome', 'SpendingScore']])
    if len(set(dbscan.labels_)) == 1:
        continue
    sscore = silhouette_score(df[['AanualIncome', 'SpendingScore']], df.cluster)
    ssamples = silhouette_samples(df[['AanualIncome', 'SpendingScore']], df.cluster)
    ylow = 10
    for c in set(dbscan.labels_):
        csample = ssamples[df.cluster==c]
        csample.sort()
        yup = ylow+csample.shape[0]
        plt.fill_betweenx(np.arange(ylow, yup), 0, csample)
        ylow = yup + 10
    plt.title(f'Score: {sscore: .2f} | eps: {eps}')
    plt.show()
```

## Parameter Space

```python
eps = np.linspace(0.1, 0.7, 10)
min_samples = np.arange(5, 15)
```

```python
parameter_space = [(e, m) for m in min_samples for e in eps]
```

```python
len(parameter_space)
```

```python
for eps, m in parameter_space:
    dbscan = DBSCAN(eps=eps, min_samples=m)
    df['cluster'] = dbscan.fit_predict(df[['AanualIncome', 'SpendingScore']])
    if len(set(dbscan.labels_)) == 1:
        continue
    sscore = silhouette_score(df[['AanualIncome', 'SpendingScore']], df.cluster) 
    if sscore < 0.4:
        continue
    ssamples = silhouette_samples(df[['AanualIncome', 'SpendingScore']], df.cluster)
    ylow = 10
    for c in set(dbscan.labels_):
        csample = ssamples[df.cluster==c]
        csample.sort()
        yup = ylow+csample.shape[0]
        plt.fill_betweenx(np.arange(ylow, yup), 0, csample)
        ylow = yup + 10
    plt.title(f'Score: {sscore: .2f} | eps: {eps}, min_samples: {m}')
    plt.show()
```

# another example

```python
# Generating sample data from make_blobs
# This particular setting has one distinct cluster and 3 clusters placed close together
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=500,
                  n_features=2,
                  centers=4,
                  cluster_std=1,
                  center_box=(-10.0, 10.0),
                  shuffle=True,
                  random_state=1)  # For reproducibility
range_n_clusters = [2, 3, 4, 5, 6]
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++',random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clusterer = KMeans(n_clusters=4, random_state=10)
cluster_labels = clusterer.fit_predict(X)
print(cluster_labels)
```

```python
import matplotlib.cm as cm 
for n_clusters in range_n_clusters:
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # 1st subplot is silhouette plot
    # silhouette coefficient can range from -1, 1 but in this example all lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # (n_clusters+1)*10 is for inserting blank space between silhouette plots of individual clusters to demarcate them clearly
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Initialize clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    # silhouette_score gives the average value for all samples
    # This gives a perspective into density and separation of formed clusters
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters =", n_clusters, "The average silhouette_score is :", silhouette_avg)

    # Compute silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate silhouette scores for samples belonging to cluster i and sort them
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        ith_cluster_silhouette_values.sort()
        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label silhouette plots with their cluster numbers at middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # vertical line for average silhouette score of all values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")
    ax1.set_yticks([])  # Clear yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing actual clusters formed
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(X[:, 0], X[:, 1], marker='.', s=30, lw=0, alpha=0.7,c=colors, edgecolor='k')

    # Labeling clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o',
                c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                    s=50, edgecolor='k')

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),fontsize=14, fontweight='bold')

plt.show()
```
