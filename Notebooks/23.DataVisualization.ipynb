```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ipywidgets helps to view interactive plot using matplotlib
import ipywidgets
```

```python
# matplotlib notebook shows visualization in interactive way, only works on default jupyter
#%matplotlib notebook
```

```python
x = np.arange(1, 10)
y = np.tan(x)
z = np.sin(y)
```

```python
#help(plt.plot)
#plt.plot?
```

### Line Plot 2D

```python
# matplotlib inline shows visualization in non interactive way
%matplotlib inline
plt.figure(num=1, figsize=(5, 3))
plt.plot(x,y,label='Tan Graph') # have to write plt.legend() to show label
plt.title('x vs tan(x)')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.legend()
plt.show()
```

```python
# matplotlib widget shows visualization in interactive way, have to declare before every plot 
%matplotlib widget
plt.figure(num=1, figsize=(5, 3))
plt.plot(x,y,label='Tan Graph') # have to write plt.legend() to show label
plt.title('x vs tan(x)')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.legend()
plt.show()
```

```python
%matplotlib widget
plt.figure(num=1, figsize=(5, 3))
plt.plot(x, y, label='Tan Graph', color='lightblue', linewidth=4, marker='o', markersize=15, markeredgecolor='black')
plt.plot(x, z, label='Sine Graph', color='lightgreen', linewidth=3, marker='^', markersize=15, markeredgecolor='gray')
plt.title('x vs Waves')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.legend()
plt.grid(True) # to show grid in plot 
plt.show()
```

```python
# to limit graph in x and y axis
%matplotlib widget
plt.figure(num=1, figsize=(5, 3))
plt.plot(x, y, label='Tan Graph', color='lightblue', linewidth=4, marker='o', markersize=15, markeredgecolor='black')
plt.plot(x, z, label='Sine Graph', color='lightgreen', linewidth=3, marker='^', markersize=15, markeredgecolor='gray')
plt.title('x vs Waves')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.legend()
plt.grid(True) # to show grid in plot 
plt.xlim([1, 9])
plt.ylim([-6, 0])
plt.show()
```

```python
# save plot as jpg, have to use savefig function before plt.show()
%matplotlib widget
plt.figure(num=1, figsize=(5, 3), dpi=100)
plt.plot(x, y, label='Tan Graph', color='lightblue', linewidth=4, marker='o', markersize=15, markeredgecolor='black')
plt.plot(x, z, label='Sine Graph', color='lightgreen', linewidth=3, marker='^', markersize=15, markeredgecolor='gray')
plt.title('x vs Waves')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.legend()
plt.grid(True)
plt.xlim([1, 9])
plt.ylim([-6, 0])
#plt.savefig('fig1.jpg', dpi=200)   
plt.show()
```

### Histogram

```python
%matplotlib widget
data = y*z
plt.figure(num=1, figsize=(5,3))
plt.hist(data, bins=5, label='Tan Distribution', orientation='vertical', color='lightblue')
plt.title('x vs Waves')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
%matplotlib widget
plt.figure(num=1, figsize=(5,3))
plt.hist(z, bins=7, label='Sine Distribution', stacked=True, orientation='horizontal', color='darkorange')
plt.title('x vs Waves')
plt.xlabel('x')
plt.ylabel('sin(y)')
plt.legend()
plt.grid(True)
plt.show()
```

### Bar Chart

```python
%matplotlib widget
langs = ['Python', 'Java', 'C++', 'JavaScript']
ratings = [8.9, 8.2, 6.9, 7.7]
plt.figure(num=1, figsize=(5, 3))
plt.bar(langs, ratings, label='Language Ratings', color='lightblue')
plt.title('Language Ratings')
plt.xlabel('Languages')
plt.ylabel('Ratings')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# using plt.barh() to plot bar chart in horizontal
%matplotlib widget
langs = ['Python', 'Java', 'C++', 'JavaScript']
ratings = [8.9, 8.2, 6.9, 7.7]
plt.figure(num=1, figsize=(5, 3))
plt.barh(langs, ratings, label='Language Ratings', color='lightblue')
plt.title('Language Ratings')
plt.xlabel('Languages')
plt.ylabel('Ratings')
plt.legend()
plt.grid(True)
plt.show()
```

### Pie Chart

```python
%matplotlib widget
langs = ['Python', 'Java', 'C++', 'JavaScript']
speed = [200, 177, 128, 113]
plt.figure(num=1, figsize=(5, 3))
plt.pie(speed, labels=langs)
plt.title('Language speed')
plt.xlabel('Languages')
plt.ylabel('speed')
plt.legend()
plt.show()
```

```python
%matplotlib widget
langs = ['Python', 'Java', 'C++', 'JavaScript']
market = [200, 177, 128, 113]
plt.figure(num=1, figsize=(5, 3))
plt.pie(market, labels=langs,autopct="%.2f")
plt.title('Market Share')
plt.show()
```

```python
%matplotlib widget
langs = ['Python', 'Java', 'C++', 'JavaScript']
market = [200, 177, 128, 113]
plt.figure(num=1, figsize=(5, 3))
plt.pie(market, labels=langs, 
        autopct="%.2f", explode=(0,0,0,0.2), colors=('lightgreen', 'gray', 'lightblue', 'red'),
        shadow=True)
plt.title('Market Share')
plt.show()
```

### Scatter plot

```python
%matplotlib widget
plt.figure(num=1, figsize=(5, 3))
plt.scatter(x, y, label='Tan Graph', color='lightblue', linewidth=4, marker='o')
plt.scatter(x, z, label='Sine Graph', color='lightgreen', linewidth=3, marker='^')
plt.title('x vs Waves')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.legend()
plt.grid(True)
plt.show()
```

### 3D

```python
%matplotlib widget
plt.figure(num=1, figsize=(5, 3))
axes = plt.axes(projection='3d')
axes.plot3D(x, y, z, label='XYZ')
axes.plot3D(x, y, z**2, label='XYZ^2')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_zlabel('Z')
plt.legend()
plt.show()
```

```python
%matplotlib widget
plt.figure(num=1, figsize=(5, 3))
axes = plt.axes(projection='3d')
axes.scatter3D(x, y, z, label='XYZ')
axes.scatter3D(x, y, z**2, label='XYZ^2')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_zlabel('Z')
plt.legend()
plt.show()
```

### Multiplots

```python
%matplotlib widget
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0,0].plot(x, y)
axes[0,0].set_title('Line Plot')
axes[0,1].scatter(x, y)
axes[0,1].set_title('Scatter Plot')
axes[1,0].hist(data)
axes[1,0].set_title('Histogram Plot')
axes[1,1].bar(langs, ratings)
axes[1,1].set_title('Bar Plot')
plt.tight_layout()
```
