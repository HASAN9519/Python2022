```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# matplotlib inline shows visualization in non interactive way
%matplotlib inline
```

```python
# matplotlib notebook shows visualization in interactive way, only works on default jupyter
#%matplotlib notebook
```

```python
x = np.arange(1, 10)
y = np.tan(x)
z = np.sin(y)
```

```python
#help(plt.plot)
#plt.plot?
```

### Line Plot 2D

```python
plt.figure(num=1, figsize=(8, 5))
plt.plot(x,y,label='Tan Graph') # have to write plt.legend() to show label
plt.title('x vs tan(x)')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.legend()
plt.show()
```

```python
plt.figure(num=1, figsize=(8, 5))
plt.plot(x, y, label='Tan Graph', color='lightblue', linewidth=4, marker='o', markersize=15, markeredgecolor='black')
plt.plot(x, z, label='Sine Graph', color='lightgreen', linewidth=3, marker='^', markersize=15, markeredgecolor='gray')
plt.title('x vs Waves')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.legend()
plt.grid(True) # to show grid in plot 
plt.show()
```

```python
# to limit graph in x and y axis
plt.figure(num=1, figsize=(8, 5))
plt.plot(x, y, label='Tan Graph', color='lightblue', linewidth=4, marker='o', markersize=15, markeredgecolor='black')
plt.plot(x, z, label='Sine Graph', color='lightgreen', linewidth=3, marker='^', markersize=15, markeredgecolor='gray')
plt.title('x vs Waves')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.legend()
plt.grid(True) # to show grid in plot 
plt.xlim([1, 9])
plt.ylim([-6, 0])
#plt.savefig('fig1.jpg', dpi=200)   
plt.show()
```

```python
plt.figure(num=1, figsize=(8, 5),dpi=100)
plt.plot(x, y, label='Tan Graph', color='lightblue', linewidth=4, marker='o', markersize=15, markeredgecolor='black')
plt.plot(x, z, label='Sine Graph', color='lightgreen', linewidth=3, marker='^', markersize=15, markeredgecolor='gray')
plt.title('x vs Waves')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.legend()
plt.grid(True)
plt.xlim([1, 9])
plt.ylim([-6, 0])
#plt.savefig('fig1.jpg', dpi=200)   
plt.show()
```

## Histogram

```python
data = y*z
plt.figure(num=1, figsize=(8, 5))
plt.hist(data, bins=5, label='Tan Distribution', orientation='vertical', color='lightblue')
# plt.hist(z, bins=7, label='Sine Distribution', stacked=True, orientation='horizontal', color='darkorange')
plt.title('x vs Waves')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.legend()
plt.grid(True)
plt.show()
```

## Bar

```python
langs = ['Python', 'Java', 'C++', 'JavaScript']
ratings = [8.9, 8.2, 6.9, 7.7]
plt.figure(num=1, figsize=(8, 5))
plt.bar(langs, ratings, label='Language Ratings', color='lightblue')
plt.title('Language Ratings')
plt.xlabel('Languages')
plt.ylabel('Ratings')
plt.legend()
plt.grid(True)
plt.show()
```

```python
langs = ['Python', 'Java', 'C++', 'JavaScript']
ratings = [8.9, 8.2, 6.9, 7.7]
plt.figure(num=1, figsize=(8, 5))
plt.barh(langs, ratings, label='Language Ratings', color='lightblue')
plt.title('Language Ratings')
plt.xlabel('Languages')
plt.ylabel('Ratings')
plt.legend()
plt.grid(True)
plt.show()
```

```python
langs = ['Python', 'Java', 'C++', 'JavaScript']
market = [200, 177, 128, 113]
plt.figure(num=1, figsize=(8, 5))
plt.pie(market, labels=langs, 
        autopct="%.2f", explode=(0,0,0,0.1), colors=('lightgreen', 'gray', 'lightblue', 'darkorange'),
        shadow=True
       )
plt.title('Market Share')
# plt.legend()
plt.grid(True)
plt.show()
```

## Scatter plot

```python
plt.figure(num=1, figsize=(8, 5))
plt.scatter(x, y, label='Tan Graph', color='lightblue', linewidth=4, marker='o')
plt.scatter(x, z, label='Sine Graph', color='lightgreen', linewidth=3, marker='^')
plt.title('x vs Waves')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.legend()
plt.grid(True)
plt.show()
```

## 3D

```python
axes = plt.axes(projection='3d')
axes.plot3D(x, y, z, label='XYZ')
axes.plot3D(x, y, z**2, label='XYZ^2')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_zlabel('Z')
plt.legend()
plt.show()
```

```python
axes = plt.axes(projection='3d')
axes.scatter3D(x, y, z, label='XYZ')
axes.scatter3D(x, y, z**2, label='XYZ^2')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_zlabel('Z')
plt.legend()
plt.show()
```

## Live PLotting

```python
x, y = [], []
for i in range(100):
    x.append(np.random.randint(1,100))
    y.append(np.random.randint(1,100))
    plt.scatter(x, y, color='lightblue')
    plt.grid(True)
    plt.pause(0.3)
plt.show()
```
