# Matplotlib
Matplotlib is a plotting library for Python programming language and its numerical mathematics extension NumPy, It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+, Some of the major Pros of Matplotlib are:
- Generally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every element in a figure
- High-quality output in many formats
- Very customizable in general

```python
import pandas as pd 
import numpy as np  
import matplotlib.pyplot as plt
%matplotlib inline
```

## scatter plot

```python
x=np.arange(0,10)
y=np.arange(11,21)
plt.scatter(x,y, c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

## line plot

```python
y=x*x
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

## Subplots

```python
# Creating Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
x = np.arange(1,11) 
y = 3 * x + 5 
plt.title("Matplotlib demo") 
plt.xlabel("x axis caption") 
plt.ylabel("y axis caption") 
plt.plot(x,y) 
plt.show()
```

```python
# Compute x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 
# Plot points using matplotlib 
plt.plot(x, y) 
plt.show() 
```

```python
# Subplot()
# Compute x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
# Set up a subplot grid that has height 2 and width 1 and set first such subplot as active
plt.subplot(2, 1, 1)
   
# Make first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set second subplot as active and make second plot 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show figure
plt.show()
```

## Bar plot

```python
# Bar plot
x = [2,8,10] 
y = [11,16,9]  
x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  
plt.show()
```

## Histograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
```

## Box Plot Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);  
```

## Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=False)
plt.axis('equal')
plt.show()
```

# Seaborn

```python
import seaborn as sns
df=sns.load_dataset("tips")
df.head()
```

## Correlation with Heatmap
A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types, It is very important in Feature Selection

```python
sns.heatmap(df.corr())
```

## JoinPlot
A join plot allows to study relationship between 2 numeric variables, central chart display their correlation, It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot, it's a bivariate Analysis

```python
# darker color means greater relation
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## Pair plot
A 'pairs plot' is also known as a scatterplot in which one variable in same data row is matched with another variable's value, Pairs plots are just elaborations on this showing all variables paired with all other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

## Displot
Displot helps us to check distribution of columns feature

```python
sns.displot(df['tip'])
```

```python
sns.displot(df['tip'],kind="kde")
```

## Categorical Plots
Seaborn also helps us in doing analysis on Categorical Data points
- boxplot
- violinplot
- countplot
- bar plot

```python
# Count plot
sns.countplot(y='sex',data=df)
```

```python
# Bar plot
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
# Bar plot
sns.barplot(x='sex',y='total_bill',data=df)
```

## Box plot
A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary

```python
sns.boxplot(x='smoker',y='total_bill', data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# categorize data based on some other categories
sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

## Violin Plot
Violin plot helps to see both distribution of data in terms of Kernel density estimation and box plot

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```

# Univariate, Bivariate and MultiVariate Analysis

```python
df=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
df.head()
```

## Univariate Analysis

```python
df_setosa=df.loc[df['species']=='setosa']
df_virginica=df.loc[df['species']=='virginica']
df_versicolor=df.loc[df['species']=='versicolor']
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
plt.xlabel('Petal length')
plt.show()
```

## Bivariate Analysis

```python
sns.FacetGrid(df,hue="species").map(plt.scatter,"petal_length","sepal_width").add_legend();
plt.show()
```

## Multivariate Analysis

```python
sns.pairplot(df,hue="species")
```
